# detailed evolution of example systems

```python
#################
def plot_variations(detail, xlim = (3.5,5.5), m1lim = (0,25),m2lim = (10,70), show_all_types = True, savestr=''):
    # stellar type definittions for plotting
    types = Table()
#     types['stellar_types'] = ['MS_low', 'MS' ,'HG', 'AGB', 'CHeB', 'EAGB','TPAGB','HeMS','HeHG','HeGB','HeWD','COWD','ONeWD','NS','BH','massless remnant']
    types['stellar_types'] = ['MS_low', 'MS' ,'HG', 'AGB', 'CHeB', 'EAGB','TPAGB','HeMS',' ','HeGB','HeWD','COWD','ONeWD','NS','BH','massless remnant']
    types['indices'] = np.arange(len(types))

    colors = ['#ca0020', '#0571b0']#sns.color_palette("coolwarm", 2)

    #########################################################
    ## Start plotting
    fig, ax = plt.subplots(figsize = (15,10), sharex=True)
    plt.subplots_adjust(hspace=0.1)
    
    #########################################################
    for mass_key in range(2):
        m = mass_key + 1
        ###################
        ### Find points where the star changes Type
        #Boolean array of when star 1 changes type
        type_change        = np.zeros(len(detail))
        type_change[1:]    = detail['Stellar_Type('+str(m)+')'][1:] - detail['Stellar_Type('+str(m)+')'][:-1] > 0
        type_change_index1 = list(np.arange(len(detail))[type_change == 1])

        ###################
        # Ax1 mass vs time
        ###################
        ### plot mass of BH with marker at every type changge
        ax.plot(detail['Time'], detail['Mass('+str(m)+')'], c = colors[mass_key], lw = 3, \
                 marker = 'o', linestyle = '--', label = '$M_{\star,%s}$'%(m), markevery = type_change_index1 , ms = 10, zorder=20) 

        ax.plot(detail['Time'], detail['Mass('+str(m)+')'], c = colors[mass_key], lw = 3, \
             marker = 'o', linestyle = '--',  markevery = type_change_index1 , ms = 10, zorder=20) 

        ###################
        ### Annotate stellar type
        # annoate most type changes for the first and last system
        for i in type_change_index1:
            if detail['Stellar_Type('+str(m)+')'][i] == 14:
                continue
            ax.annotate("$\mathrm{%s}$"%(types['stellar_types'][detail['Stellar_Type('+str(m)+')'][i]]),
                        (detail['Time'][i] , detail['Mass('+str(m)+')'][i]),annotation_clip=False,
                        size=21, color= colors[mass_key], rotation=20, va = 'bottom', zorder=2)#.draggable()
        # Annotate BH slightly lower
        ax.annotate("$\mathrm{%s}$"%(types['stellar_types'][detail['Stellar_Type('+str(m)+')'][i]]),
            (detail['Time'][i] , detail['Mass('+str(m)+')'][i]-1),annotation_clip=False,
            size=21, color= colors[mass_key], rotation=20, va = 'top', zorder=2)

    ##################
    # plot values
    ### ax0 M1 vs time ###
    ax.set_xlim(xlim)
    ax.set_ylim(m1lim)
    ax.set_xlabel('$\mathrm{time [Myr]}$', size = 30)
    ax.set_ylabel('$M_1 \ \mathrm{[M_{\odot}]}$', size = 35)  
    ax.xaxis.grid(5) # vertical lines

    #################
    # second axis
    ax2 = ax.twinx()
    ax2.plot(detail['Time'], detail['SemiMajorAxis'], c = '#fdae61', 
             lw = 5, alpha =0.7, linestyle = '-', zorder =0) 
    ax2.set_ylabel('$\mathrm{Separation \ [R_{\odot}]}$', size = 35)  
    
    ax.legend(loc = 'lower left')
    print('saving here:',save_loc +'/Detail'+savestr+'.pdf' )
    plt.savefig(save_loc +'/Detail'+savestr+'.pdf', bbox_inches='tight')
    plt.savefig(save_loc +'/png_versions/Detail'+savestr+'.png', bbox_inches='tight')
    plt.show()

    return 



######### Read the system ##############
seed = 29118#16103#17460#15127

data_dir    = '/Users/lieke/surfdrive/Documents/CompareCOMPAS/CompasOutput/vdHeuvelSystem/'
input_dir_detailed = data_dir +str(seed)+'/Detailed_Output/BSE_Detailed_Output_0.csv'
    
print('\n'+10*'*', seed, 10*'*')
DCO = Table.read(data_dir +str(seed)+'/BSE_Double_Compact_Objects.csv', header_start=2,data_start=3, format='csv', delimiter = ',')
display(DCO)
print(DCO['Coalescence_Time']*1e-3, 'Gyr')

detail = Table.read(input_dir_detailed, header_start=2,data_start=3, format='csv', delimiter = ',')

######### plot it ##############
plot_variations(detail, xlim = (3,6.), m1lim = (0,150), m2lim = (10,70), 
                show_all_types = True,savestr=str(seed) )



```
