# Read your data

the function `read_data` is part of `HelperFunctions.py` and reads the hdf5 file containing the BBH population data and merger rates. This is expensive memory-wise
Hence the columns specifically needed for this  Figure are loaded right below that

The Bool "DCO_mask" filters for BBHs:  
1. with an inspiral time that is less than the age of the Universe
2. excludes systems that experienced a CE from a HG donor (i.e. the flag `Optimistic_CE == False`)
3. excludes systems that experienced RLOF immediately following a CE (i.e. the flag `Immediate_RLOF>CE == False`)

In other words, we treat 2. and 3. as stellar mergers and exclude them from the rest of our analysis

Lastly, we select merging BBHs using the `DCO_mask`, and aditionally exclude systems that evolve Chemically homogeneous. 


```python
####################################################
## Location of your data
File_location = data_dir+'COMPAS_Output_wWeights.h5'
print(File_location)

# Rate selects wich mergers you are interested in, using a specific SFRD
rate_key      = 'Rates_mu00.025_muz-0.05_alpha-1.77_sigma01.125_sigmaz0.05_zBinned'

####################################################
'''
The following function will read more information (generally useful for all plots in paper)
Warning, this is heavy on memory usage 
Bare essential data will be read below
DCO, DCO_mask, redshifts, Average_SF_mass_needed, intrinsic_rate_density, intrinsic_rate_density_z0, = func.read_data(loc = File_location, rate_key = rate_key)
DCO.info()
'''

################################################
## Essential data for this plot
## Open hdf5 file and read relevant columns
File        = h5.File(File_location ,'r')

DCO = Table()
DCO['SEED']                  = File['DoubleCompactObjects']['SEED'][()] 
# DCO['Metallicity@ZAMS(1)']   = File['DoubleCompactObjects']['Metallicity@ZAMS(1)'][()] 
DCO['CE_Event_Count']        = File['DoubleCompactObjects']['CE_Event_Count'][()] 
DCO['M_moreMassive']         = np.maximum(File['DoubleCompactObjects']['Mass(1)'][()], File['DoubleCompactObjects']['Mass(2)'][()])
# DCO['mixture_weight']        = File['DoubleCompactObjects']['mixture_weight'][()]

SYS_DCO_seeds_bool           = np.in1d(File['SystemParameters']['SEED'][()], DCO['SEED']) #Bool to point SYS to DCO
DCO['Stellar_Type@ZAMS(1)']  = File['SystemParameters']['Stellar_Type@ZAMS(1)'][SYS_DCO_seeds_bool]

DCO_mask                     = File[rate_key]['DCOmask'][()] # Mask from DCO to merging BBH 
intrinsic_rate_density       = File[rate_key]['merger_rate'][()]
redshifts                    = File[rate_key]['redshifts'][()]

File.close()

################################################
# Bools to select merging BBHs w.o. CHE only
nonCHE_bool         = DCO['Stellar_Type@ZAMS(1)'] != 16
rate_nonCHE_bool    = DCO['Stellar_Type@ZAMS(1)'][DCO_mask] != 16

# Filter both the BBH table and the intrinsic rate data
merging_BBH         = DCO[DCO_mask * nonCHE_bool]
Red_intr_rate_dens  = intrinsic_rate_density[rate_nonCHE_bool, :]
```

# Average merger rate density in Mbh-z plane

```python
##############################
# plot function
##############################
def plot_rate_dens_Mbh_z(table, intrinsic_rate_density, y_key = 'M_moreMassive',
                         Crude_z_binsize = 0.1, tmin=0, tmax = 10, x_redshift=True,
                          Vmin =1e-2, Vmax=20, contour_levels = [0.01, 0.1, 1.,2.5, 10],
                         ylabel = '$M_{\mathrm{BH,1}}$ $\mathrm{[M_{\odot}]}$', 
                         colormap = sns.color_palette("rocket", as_cmap=True),
                         title_str = ' All', save_plot=False, savestr='All'):

    """
    Plot 2D histogram of average rate density in Mbh-redshift plane 


    Args:
            table                   --> [astropy table] table containing your merging BBHs
            intrinsic_rate_density  --> [2D array of floats] the intrinsic merger rate of your BBHs
            y_key                   --> [string] what column to put on the y axis
            Crude_z_binsize         --> [float] how big do you want your redshift bins
            tmin, tmax              --> [float, float] set min and max in lookback time 
            x_redshift              --> [bool] if you want redshifts on your xaxis (alternative is lookbacktime)
            Vmin, Vmax              --> [float, float] min and max of colour values
            contour_levels          --> [float array] where to draw contours
            ylabel                  --> [string] what to put on the yaxis
            colormap                --> [cmap] matplotlib cmap used for the histogram shading
            title_str               --> [string] title that will be printed on the top left your plot
            save_plot               --> [bool] if you would like to save your plot or not
            savestr                 --> [string] will be appended in front of your image filename

        Returns:
            matplotlib figure  -->  Figure of 2D histogram

    """
    ##############################
    fig, ax = plt.subplots(figsize = (11, 7))    
    
    ##############################
    #Define your crude redshift bins
    Crude_redshifts = np.arange(0, 10+ Crude_z_binsize, Crude_z_binsize)
    center_Crude_bins = (Crude_redshifts[:-1] + Crude_redshifts[1:])/2. # center points

    ##############################
    ## Calculate average rate density per z-bin
    crude_rate_density = func.get_crude_rate_density(intrinsic_rate_density, redshifts, Crude_redshifts )

    ##############################
    # Define your mass bins
    mass_bins = np.arange(0, 50, 2.5)
    center_mass_bin =  (mass_bins[:-1] + mass_bins[1:])/2. # center points
    mass_binsizes    = np.diff(mass_bins)

    ##############################
    # Repeat the redshifts, Mass times in vertical direction
    X = np.repeat(center_Crude_bins[np.newaxis,:], len(table[y_key]), axis=0 )
    # Repeat the masses, z times in horizontal direction
    Y = np.repeat(table[y_key][:, np.newaxis], len(center_Crude_bins), axis=1)
    #X and Y now have the same shape as crude_rate_density


    ##############################
    ## Hist the flat version of all your points
    counts, redshift_edges, mass_edges = np.histogram2d(X.ravel(), Y.ravel(), bins=[Crude_redshifts, mass_bins],  
                                                        density=False, weights=crude_rate_density.ravel())  
    # Divide by dM
    Z_histvals = counts/mass_binsizes
    
    rate_per_mass = np.sum(Z_histvals.T, axis = 1)
    
    # Plot the colourmesh
    Rcolor = ax.pcolormesh(center_Crude_bins, center_mass_bin, Z_histvals.T, 
                           norm=matplotlib.colors.LogNorm(vmin=Vmin,vmax=Vmax), 
                           rasterized=True, cmap=colormap, alpha=0.95 , shading='auto') #YlGnBu
    cb = plt.colorbar(Rcolor, label= "$d\overline{R}_{\mathrm{BBH}}/dM_{\mathrm{BBH,1}} \ \mathrm{[Gpc^{-3}yr^{-1} M_{\odot}^{-1}]} $")  

    ###### Contour from hist ######
    contour_kleur = sns.light_palette("#000000", as_cmap=True)
    CS = plt.contour(Z_histvals.T, extent=[redshift_edges.min(),redshift_edges.max(),mass_edges.min(),mass_edges.max()],
                     linewidths=3, levels=contour_levels, cmap=contour_kleur,#'Greys_r',
                     norm=matplotlib.colors.LogNorm(vmin = 1e-3, vmax = np.amax(Z_histvals)))
    
    ax.clabel(CS, inline=1, fontsize=20,fmt='%1.2f')    

    ##########################################################################
    #### Age Universe Axis ####
    ax2 = ax.twiny()

    if x_redshift:
        ###################
        #Plotvalues
        ax.set_xlabel('$\mathrm{merger \ redshift}, \ z$', fontsize = 30)

        # Find lookback location for each of our redshifts
        redshift_tick_list = [0, 2, 4, 6, 8, 10]
        z_firstSF = 10 # Start counting from here (should be the same as in your merger data)
        t_firstSF = cosmo.lookback_time(z_firstSF).value
        # And annotate the tick labels :)
        ax2.set_xticks([z for z in redshift_tick_list])
        ax2.set_xticklabels(['${:.1f}$'.format(t_firstSF - cosmo.lookback_time(z).value) for z in redshift_tick_list])
        ax2.set_xlabel('$\mathrm{time \ since} \ z=10 \ \mathrm{[Gyr]}$', fontsize = 30, labelpad=15)

    else:
        ###################
        #Plotvalues
        ax.set_xlabel(r'$\mathrm{Lookback \ time \ [Gyr]}$', fontsize = 30)
        
        redshift_tick_list = [0,0.1, 0.25, 0.5, 0.75, 1.0,1.5, 2, 3, 6, ]
        # Find lookback location for each of our ages
        z_ticks = [cosmo.lookback_time(z) for z in redshift_tick_list]

        # And annotate the tick labels :)
        ax2.set_xticks([cosmo.lookback_time(z).value for z in redshift_tick_list])
        ax2.set_xticklabels(['${:g}$'.format(z) for z in redshift_tick_list])
        ax2.set_xlabel('$\mathrm{merger \ redshift}, \ z$', fontsize = 30)
    
        
    # Make sure top and bottom axis are lined up (have same limmits)
    ax.set_xlim(tmin, tmax)
    ax2.set_xlim(tmin, tmax)
    ##########################################################################
 
    ########################
    # Plot values
    ax.set_facecolor('k')
    
    plt.ylim(0,55)
    ax.set_ylabel(ylabel, fontsize=30)
    ax.text(0.3, 52.5, s=title_str, ha='left', va='top', fontsize = 30,
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.5', alpha = 0.7))

    if save_plot:
        print('saving here: ', save_loc+'/'+savestr+'Mass_redshift_'+y_key+'.pdf')
        plt.savefig(save_loc+'/'+savestr+'Mass_redshift_'+y_key+'.pdf', bbox_inches='tight')
        plt.savefig(save_loc+'/png_versions/'+savestr+'Mass_redshift_'+y_key+'.png', bbox_inches='tight')

    plt.show()
    
    return rate_per_mass
    
    
    
```

```python
########################################
## All BBH
Bool =  DCO_mask *nonCHE_bool
plot_rate_dens_Mbh_z(merging_BBH, Red_intr_rate_dens, 
                     title_str = '$\mathrm{All \ BBHs}$', savestr=sim_save_str+'All', save_plot=True)

```
